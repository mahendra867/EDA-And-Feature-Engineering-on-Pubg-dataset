## Solos, Duos and Squads

#### There are 3 game modes in the game. One can play solo, or with a friend (duo), or with 3 other friends (squad). 100 players join the same server, so in the case of duos the max teams are 50 and in the case of squads the max teams are 25

```python
solos=train[train['numGroups']>50] # The code you provided is creating three subsets of the train dataframe based on the number of groups in each game.,The train dataframe seems to be related to a multiplayer online game, such as PlayerUnknown's Battlegrounds (PUBG), where players are divided into groups and compete against each other. The 'numGroups' column in the dataframe holds the number of groups in each game. The first subset, solos, selects all games where the number of groups is greater than 50. These games are assumed to be solo games, where each player competes individually without being part of a group.
duos=train[(train['numGroups']>25)& (train['numGroups']<=50)] # The second subset, duos, selects all games where the number of groups is greater than 25 but less than or equal to 50. These games are assumed to be duo games, where each player competes as part of a two-person group.
squads=train[train['numGroups']<=25] # The third subset, squads, selects all games where the number of groups is less than or equal to 25. These games are assumed to be squad games, where each player competes as part of a group of more than two people.
print("There are {} ({:.2f}%) solo games, {} ({:.2f}%) duo games and {} ({:.2f}%) squad games.".format(len(solos), 100*len(solos)/len(train), len(duos), 100*len(duos)/len(train), len(squads), 100*len(squads)/len(train),))# The len() function is used to calculate the number of games in each subset. These values are then used to calculate the percentage of games in each subset relative to the total number of games in the train dataframe. Finally, the print() function is used to display the results in a formatted string, which reports the number and percentage of solo, duo, and squad games. The format() method is used to insert the values of the variables into the string.
```

```python
f,ax1=plt.subplots(figsize=(20,10))
sns.pointplot(x='kills',y='winPlacePerc',data=solos,color='black')
sns.pointplot(x='kills',y='winPlacePerc',data=duals,color='#CC0000')
sns.pointplot(x='kills',y='winPlacePerc',data=squads,color='#3399FF')
plt.text(37,0.6,'solos',color='black',fontsize=17,style='italic')
plt.text(37,0.55,'duals',color='#CC0000',fontsize=17,style='italic')
plt.text(37,0.5,'squads',color='#3399FF',fontsize=17,style='italic')
plt.xlabel('Number of Kills',fontsize=15,color='blue')
plt.xlabel('win percentage',fontsize=15,color='blue')
plt.title('Solo vs Duo vs Squad Kills',fontsize=20,color='blue')
plt.grid()
plt.show()
```

```python
train['playersJoined'] = train.groupby('matchId')['matchId'].transform('count') # This code creates a new column called 'playersJoined' in the train DataFrame. The values of this column represent the number of players who participated in each match. The calculation is performed using the groupby() method of the DataFrame, which groups the rows by the 'matchId' column. The transform() method is then called on the resulting groupby object with the argument 'count', which counts the number of rows in each group. This results in a new Series object with the same length as the original DataFrame, containing the count of players for each match. Finally, the resulting Series is assigned to the 'playersJoined' column of the train DataFrame, which adds a new column to the DataFrame with the count of players for each match. Overall, this code allows us to add a new column to the train DataFrame that provides information about the number of players in each match, which can be useful for analyzing the data and building models.


#The difference between ('matchId') and ['matchId'] is that the former is a string enclosed in parentheses, while the latter is a list containing a string. In pandas, when using the square bracket notation to select columns from a DataFrame, the column names should be passed as a list of strings. For example, df[['col1', 'col2']] selects the 'col1' and 'col2' columns from the DataFrame df. On the other hand, when using the parentheses notation to call a method on a DataFrame, the arguments should be enclosed in parentheses. For example, df.groupby('col1') groups the rows of the DataFrame df by the values in the 'col1' column.
```

#### Based on the "playersJoined" feature we can create (or change) a lot of others to normalize their values. For example i will create the "killsNorm" and "damageDealtNorm" features. When there are 100 players in the game it might be easier to find and kill someone, than when there are 90 players. So i will normalize the kills in a way that a kill in 100 players will score 1 (as it is) and in 90 players it will score (100-90)/100 + 1 = 1.1. This is just an assumption. You can use different scales

```python
train['killsNorm']=train['kills']*((100-train['playersJoined'])/100+1)
train['damageDealtNorm']=train['damageDealt']*((100-train['playersJoined'])/100+1)
train[['playersJoined', 'kills', 'killsNorm', 'damageDealt', 'damageDealtNorm']][5:8] #A i have selected some column name and iam trying to fetvch the only rows which are from  5 to 7 rows  
```

#### Another simple feature is the sum of heals and boosts. Also the sum of total distance travelled.

```python
train['healsAndBoosts']=train['heals']+train['boosts'] # here i have selected the heals and boosts columns and adding them and pasting it in heals and boost column got newly created
train['totalDistance']=train['walkDistance']+train['rideDistance']+train['swimDistance'] # here i have selected walkDistance,rideDistance,swimDistance and adding them and pasting it in total distance column which got newly created ine a datafram
```

#### When using boosting items you run faster. They also help staying out of the zone (PUBG term) and loot more (meaning walking more). So lets create a feature boosts per walking distance. Heals don't make you run faster, but they also help staying out of the zone and loot more. So lets create the same feature for heals also.

```python
train['boostsPerWalkDistance'] = train['boosts']/(train['walkDistance']+1) # This line of code creates a new column in a pandas DataFrame called train called 'boostsPerWalkDistance'. The values in this column represent the ratio of the 'boosts' column to the 'walkDistance' column, with a small adjustment to avoid dividing by zero. The calculation is performed by dividing the 'boosts' column by the 'walkDistance' column plus one. The addition of one is to avoid dividing by zero in cases where 'walkDistance' is zero. This can happen in situations where a player has used a boost item but has not moved, resulting in a 'walkDistance' value of zero. In such cases, the ratio of 'boosts' to 'walkDistance' would be undefined or infinite, so adding one ensures that the ratio is well-defined and finite.
train['boostsPerWalkDistance'].fillna(0, inplace=True) # This line of code fills in missing values in the 'boostsPerWalkDistance' column of the train DataFrame with a value of zero. The fillna() method is called on the 'boostsPerWalkDistance' column of the DataFrame, and takes two arguments: 0: This is the value to be used to fill in the missing values. In this case, it is zero. inplace=True: This argument specifies that the operation should be performed on the DataFrame in place, i.e., the original DataFrame should be modified rather than creating a new copy. The reason for filling in missing values with zero is not clear from the code snippet alone, as it depends on the context of the analysis. However, it is common to fill in missing values with some default value in order to avoid errors or inconsistencies in subsequent analysis or modeling steps that require complete data. In this case, since the 'boostsPerWalkDistance' column represents a ratio of two other columns, it is possible that there are some cases where both the 'boosts' and 'walkDistance' columns are zero, resulting in a missing value in the 'boostsPerWalkDistance' column. Filling in these missing values with zero may be a reasonable choice, as it ensures that the ratio is well-defined and consistent across all rows of the DataFrame.
train['healsPerWalkDistance'] = train['heals']/(train['walkDistance']+1) #The +1 is to avoid infinity, because there are entries where heals>0 and walkDistance=0. Strange.
train['healsPerWalkDistance'].fillna(0, inplace=True)
train['healsAndBoostsPerWalkDistance'] = train['healsAndBoosts']/(train['walkDistance']+1) #The +1 is to avoid infinity.
train['healsAndBoostsPerWalkDistance'].fillna(0, inplace=True)
train[['walkDistance', 'boosts', 'boostsPerWalkDistance' ,'heals',  'healsPerWalkDistance', 'healsAndBoosts', 'healsAndBoostsPerWalkDistance']][40:45]
```
